```python
# Função para gerar o script de criação da tabela
def generate_create_table_script(df, table_name):
    create_table_script = f"CREATE TABLE {table_name} (\n"
    columns = df.columns

    for col in columns:
        dtype = df[col].dtype
        
        if dtype == 'int64':
            column_type = 'INT'
        elif dtype == 'float64':
            column_type = 'FLOAT'
        elif dtype == 'bool':
            column_type = 'BOOLEAN'
        elif dtype == 'object':
            column_type = 'TEXT'
        else:
            column_type = 'TEXT'

        create_table_script += f"    {col} {column_type},\n"

    create_table_script = create_table_script.rstrip(',\n') + "\n);"

    return create_table_script

# Nome da tabela desejada
table_name = 'rate'

# Geração do script de criação da tabela
create_table_script = generate_create_table_script(df_rate, table_name)

# Exibição do script gerado
print(create_table_script)
```

```python
def execute_create_table_script(script):
    try:
        # Conexão com o banco de dados
        cursor = conn.cursor()

        # Executa o script de criação da tabela
        cursor.execute(script)

        # Confirma as alterações no banco de dados
        conn.commit()

        # Fecha a conexão
        cursor.close()
        conn.close()

        print("Tabela criada com sucesso!")
    except (Exception, psycopg2.Error) as error:
        print("Erro ao criar a tabela:", error)

execute_create_table_script(create_table_script, )
```

```python
# Unindo os dataframes
merged_df = pd.merge(df_rate, df_plan, left_on=df_rate['RATE_PLANID'], right_on=df_plan['PLANATTRIBUTES_PLANID'].str[:14], how='inner')
```

```python
# Filtrando os dados
filtered_df = merged_df[(merged_df['PLANATTRIBUTES_DENTALONLYPLAN'] == 'Yes') & (merged_df['RATE_INDIVIDUALRATE'] != 999999)]
filtered_df
```

```python
filtered_df['RATE_STATECODE']
filtered_df['PLANATTRIBUTES_STATECODE']
filtered_df['indi']
```

```python
# Calculando a média do individualrate por statecode e businessyear
grouped_df = filtered_df.groupby(['RATE_STATECODE', 'RATE_BUSINESSYEAR']).agg({'RATE_INDIVIDUALRATE': 'mean'}).reset_index()
grouped_df
```

```python
# Ordenando por statecode
sorted_df = grouped_df.sort_values('RATE_STATECODE')
```
